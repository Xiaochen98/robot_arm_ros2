```python
#!/usr/bin/env python3
#coding=utf-8
import time
from Arm_Lib import Arm_Device

# 创建机械臂对象 Create a Robotic Arm Object
Arm = Arm_Device()
time.sleep(.1)
```

```python
# 打开学习模式，扩展板上的RGB灯呈现呼吸灯状态，同时机械臂所有舵机都进入关闭扭矩状态，
# 即可以自由掰动，可以把机械臂掰动到要记住的位置上
# Turn on the learning mode, the RGB light on the expansion board will show the breathing light state, and all the servos of the robot arm will enter the torque-off state.
# That means you can move it freely and move the robot arm to the position you want to remember
Arm.Arm_Button_Mode(1)
```

```python
# 在学习模式下，每次运行此cell，记录当前的动作保存到扩展板，同时扩展板上的RGB灯会切换颜色，
# 如果出现了红色呼吸灯，则表示学习的动作组已经满（20组）。
# 此命令也可以替换成按扩展板上的K1键，两者的功能是一致的。
# In learning mode, each time you run this cell, the current action is recorded and saved to the expansion board, and the RGB light on the expansion board will switch colors.
# If a red breathing light appears, it means that the learned action group is full (20 groups).
# This command can also be replaced by pressing the K1 key on the expansion board. The functions of the two are the same.
Arm.Arm_Action_Study()
```

```python
# 关闭学习模式。关闭呼吸灯
# Turn off learning mode. Turn off breathing light
Arm.Arm_Button_Mode(0)
```

```python
# 读取当前记录的动作组数量
# Read the number of action groups currently recorded
num = Arm.Arm_Read_Action_Num()
print(num)
```

```python
# 单次运行动作组
# Single run action group
Arm.Arm_Action_Mode(1)
```

```python
# 循环运行动作组
# Loop through the action group
Arm.Arm_Action_Mode(2)
```

```python
# 停止动作组
# Stop action group
Arm.Arm_Action_Mode(0)
```

```python
# 清空动作组，清空完成扩展板上的RGB灯会亮绿色。
# 注意：一旦清空已记录的动作组则无法恢复。
# Clear the action group. When the clearing is completed, the RGB light on the expansion board will light up green.
# Note: Once the recorded action group is cleared, it cannot be restored.
Arm.Arm_Clear_Action()
```

```python
del Arm  # 释放掉Arm对象 Release the Arm object
```
