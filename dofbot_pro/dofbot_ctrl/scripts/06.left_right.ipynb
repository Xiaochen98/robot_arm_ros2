```python
#!/usr/bin/env python3
#coding=utf-8
import time
from Arm_Lib import Arm_Device

# 创建机械臂对象 Create a Robotic Arm Object
Arm = Arm_Device()
time.sleep(.1)
```

```python
# 循环控制机械臂上下左右摆动 Cyclic control of the robot arm to swing up, down, left, and right
def main():
    # 让舵机复位归中 Reset the servo to center
    Arm.Arm_serial_servo_write6(90, 90, 90, 90, 90, 90, 500)
    time.sleep(1)


    while True:
        # 控制3号和4号舵机上下运行 Control the No. 3 and No. 4 servos to move up and down
        Arm.Arm_serial_servo_write(3, 0, 1000)
        time.sleep(.001)
        Arm.Arm_serial_servo_write(4, 180, 1000)
        time.sleep(1)
        
        # 控制1号舵机左右运动
        Arm.Arm_serial_servo_write(1, 180, 500)
        time.sleep(.5)
        Arm.Arm_serial_servo_write(1, 0, 1000)
        time.sleep(1)
        
        # 控制舵机恢复初始位置 Control the servo to restore the initial position
        Arm.Arm_serial_servo_write6(90, 90, 90, 90, 90, 90, 1000)
        time.sleep(1.5)


try :
    main()
except KeyboardInterrupt:
    print(" Program closed! ")
    pass
```

```python
del Arm  # 释放掉Arm对象 Release the Arm object
```
