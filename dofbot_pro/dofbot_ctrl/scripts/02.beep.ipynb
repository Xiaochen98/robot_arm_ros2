```python
#!/usr/bin/env python3
#coding=utf-8
import time
from Arm_Lib import Arm_Device

# 获取机械臂的对象 Get the robot arm object
Arm = Arm_Device()
time.sleep(.1)
```

```python
# 蜂鸣器自动响100毫秒后关闭
# The buzzer automatically turns off after ringing for 100 milliseconds
b_time = 1
Arm.Arm_Buzzer_On(b_time)
time.sleep(1)
```

```python
# 蜂鸣器自动响300毫秒后关闭
# The buzzer automatically turns off after ringing for 300 milliseconds
b_time = 3
Arm.Arm_Buzzer_On(b_time)
time.sleep(1)
```

```python
# 蜂鸣器一直响
# The buzzer keeps ringing
Arm.Arm_Buzzer_On()
time.sleep(1)
```

```python
# 关闭蜂鸣器
# Turn off the buzzer
Arm.Arm_Buzzer_Off()
time.sleep(1)
```

```python
del Arm  # 释放掉Arm对象 Release the Arm object
```
