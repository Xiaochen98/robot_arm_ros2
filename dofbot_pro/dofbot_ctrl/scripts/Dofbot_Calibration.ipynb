```python
#!/usr/bin/env   python3
#coding=utf-8
import   time
from   Arm_Lib import Arm_Device

# 创建机械臂对象
# Create a Robotic Arm Object
Arm   = Arm_Device()
time.sleep(.1)
```

```python
# 让舵机归中直立状态
# Let the servo return to the center upright position
Arm.Arm_serial_servo_write6(90, 90, 90, 90, 90, 180, 1000)
time.sleep(2)
```

```python
# 关闭扭矩,此时可以用手掰动舵机的角度
# Turn off the torque, and you can adjust the servo angle by hand.
Arm.Arm_serial_set_torque(0)
```

```python
# 调节好了某个舵机的角度后，可以单独设置某个舵机中位偏差
id   = 6
Arm.Arm_serial_servo_write_offset_switch(id)
time.sleep(.1)
state   = Arm.Arm_serial_servo_write_offset_state()
if   state == 1:
    print("set offset ok!")
elif   state == 2:
    print("error! set offset overrun   !")
elif   state == 0:
    print("error! set offset error   !")
```

```python
# 一次性设置全部舵机（1-6号）中位偏差
# Set the median deviation of all servos (1-6) at one time
for i in range(6):
    id = i + 1
    Arm.Arm_serial_servo_write_offset_switch(id)
    time.sleep(.1)
    state =   Arm.Arm_serial_servo_write_offset_state()
    if state == 1:
        print("id:%d set offset   ok!" % id)
    elif state == 2:
        print("error!id:%d set offset   overrun !" % id)
    elif state == 0:
        print("error!id:%d set offset   error !" % id)
```

```python
# 调整完成后，打开扭矩
# Set the median deviation of all servos (1-6) at one time
Arm.Arm_serial_set_torque(1)
```

```python
# 恢复默认状态：清除所有舵机设置的中位偏差
# Restore default state: clear the median deviation of all servo settings
# Arm.Arm_serial_servo_write_offset_switch(0)
```

```python
del Arm  # 释放掉Arm对象 Release the Arm object
```
