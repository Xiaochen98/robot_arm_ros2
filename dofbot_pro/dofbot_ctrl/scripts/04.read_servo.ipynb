```python
#!/usr/bin/env python3
#coding=utf-8
import time
from Arm_Lib import Arm_Device

# 创建机械臂对象 Create a Robotic Arm Object
Arm = Arm_Device()
time.sleep(.1)
```

```python
# 读取所有舵机的角度，并循环打印出来 Read the angles of all servos and print them out in a loop
def main():

    while True:
        for i in range(6):
            aa = Arm.Arm_serial_servo_read(i+1)
            print(aa)
            time.sleep(.01)
        time.sleep(.5)
        print(" END OF LINE! ")
        break

    
try :
    main()
except KeyboardInterrupt:
    print(" Program closed! ")
    pass
```

```python
# 单独控制一个舵机运动后，再读取它的角度 After controlling the movement of a servo individually, read its angle
id = 6
angle = 150

Arm.Arm_serial_servo_write(id, angle, 500)
time.sleep(1)

aa = Arm.Arm_serial_servo_read(id)
print(aa)

time.sleep(.5)
```

```python
del Arm  # 释放掉Arm对象 Release the Arm object
```
