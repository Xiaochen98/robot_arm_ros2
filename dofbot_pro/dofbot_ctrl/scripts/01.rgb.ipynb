```python
#!/usr/bin/env python3
#coding=utf-8
import time
from Arm_Lib import Arm_Device

# 获取机械臂的对象 Get the robot arm object
Arm = Arm_Device()
time.sleep(.1)

def main():
    while True:
        Arm.Arm_RGB_set(50, 0, 0) # RGB亮红灯 RGB lights up red
        time.sleep(.5)
        Arm.Arm_RGB_set(0, 50, 0) # RGB亮绿灯 RGB green light
        time.sleep(.5)
        Arm.Arm_RGB_set(0, 0, 50) # RGB亮蓝灯 RGB blue light
        time.sleep(.5)

        print(" END OF LINE! ")

try :
    main()
except KeyboardInterrupt:
    # 释放Arm对象 Release the Arm object
    del Arm
    print(" Program closed! ")
    pass
```
